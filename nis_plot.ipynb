```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vals=[]
with open('nis.out') as nis_data :
    for line in nis_data:
        vals.append(line.strip())
    
```

```python
fig=plt.figure(figsize=(10,6))
plt.plot(vals,label="NIS")
plt.plot([1,len(vals)],[7.815,7.815],'-',label="95%")
plt.legend(loc=1)
plt.xlabel('steps')
fig.tight_layout()
plt.show()
```
